```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1,2,3,4,5]
y=[12,14,11,15,1]
#fig,ax=plt.subplots()

plt.xlabel('x_values')
plt.ylabel('y_values')
plt.title("line plots")
plt.plot(x,y,color="purple",linestyle='dotted',linewidth=2,marker='*')
plt.show()
```

```python
months=['jan','feb','mar','apr','may','jun']
savings_amount=[9000,12880,1100,2800,8990,9500]

plt.xlabel('months')
plt.ylabel('saving amount')
plt.title("saving amount per month")
plt.plot(months,savings_amount,color="green",linestyle='dashed',linewidth=1,marker='*')
plt.show()
```

```python
categories=['A','B','C','D','E']
values=[10,15,7,12,9]
plt.bar(categories,values,color=['purple','yellow','blue','red','green'])
plt.xlabel('categories')
plt.ylabel('values')
plt.title("Basic Bar Chart")
plt.show()
```

```python
categories=['A','B','C','D','E','F']
values=[10,15,7,12,9,12]
plt.bar(categories,values,color=['orange','white','green'],edgecolor='black')

plt.xlabel('categories')
plt.ylabel('values')
plt.title("Basic Bar Chart")
plt.show()
```

```python
ages=[20,22,22,32,45,47,67,28,70,31,32,34,55,35,30]
b=[20,30,40,50,60,70]
plt.hist(ages,bins=b,color='skyblue',edgecolor='black')
plt.title('Age distribution of customers')
plt.xlabel('Age range')
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
prm_lang=['python','Java','Ruby','c++']
percen=[40,30,10,20]
plt.pie(percen,labels=prm_lang,colors=['skyblue','yellow','pink','green'],
        autopct='%1.1f%%',wedgeprops={'edgecolor':'black'})
plt.title('Programming languauge usage')
plt.show()
```

```python
x=[20,32,45,67,70,31]
y=[20,30,40,50,60,70]
plt.scatter(x,y,color='yellow',s=100,edgecolor='black')
plt.xlabel('x_values')
plt.ylabel('y_values')
plt.title("scatter plot example")

plt.show()
```
