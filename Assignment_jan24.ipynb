```python
import pandas as pd
import matplotlib.pyplot as plt
```

# 1. Sales Analysis

```python
sales_data = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [100, 500, 200, 280, 389, 320],
    'Order_Quantity': [50, 65, 55, 70, 75, 80]
})
```

# Line plot for monthly sales

```python
plt.figure(figsize=(10, 5))
plt.plot(sales_data['Month'], sales_data['Sales'], marker='o', linestyle='-', color='b')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()


```

# Histogram for order quantity distribution

```python

plt.figure(figsize=(10, 5))
plt.hist(sales_data['Order_Quantity'], bins=5, color='purple', edgecolor='black')
plt.title('Order Quantity Distribution')
plt.xlabel('Order Quantity')
plt.ylabel('Frequency')
plt.show()
```

# 2. Customer Demographics

```python
customer_data = pd.DataFrame({
    'Age': [25, 30, 22, 45, 33, 29, 40, 50, 38, 28],
    'Spending_Score': [70, 65, 80, 55, 60, 75, 58, 40, 62, 78]
})
```

# Histogram for age distribution

```python
plt.figure(figsize=(10, 5))
plt.hist(customer_data['Age'], bins=5, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

```

# Scatter plot for spending patterns

```python
plt.figure(figsize=(10, 5))
plt.scatter(customer_data['Age'], customer_data['Spending_Score'], color='purple', alpha=0.7)
plt.title('Spending Patterns by Age')
plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.grid(True)
plt.show()
```

# 3. School Performance

```python
school_data = pd.DataFrame({
    'Student': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Math': [85, 78, 92, 88, 76],
    'Science': [90, 85, 89, 94, 80],
    'English': [78, 83, 88, 82, 75]
})
```

# Bar chart for marks comparison across subjects

```python
school_data.set_index('Student')[['Math', 'Science', 'English']].plot(kind='bar', figsize=(10, 5))
plt.title('Students Performance Across Subjects')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.legend(title='Subjects')
plt.grid(axis='y')
plt.show()
```

# Box plot to identify outliers

```python
plt.figure(figsize=(10, 5))
school_data[['Math', 'Science', 'English']].plot(kind='box')
plt.title('Marks Distribution Across Subjects')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```
