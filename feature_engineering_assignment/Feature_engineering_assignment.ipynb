```python
import pandas as pd 
```

```python
df=pd.read_csv("dataset_of_employees.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
# Identifying numerical and categorical columns
numerical_cols = df.select_dtypes(include=["int64", "float64"]).columns
categorical_cols = df.select_dtypes(include=["object"]).columns
```

```python
for col in numerical_cols:
    if df[col].isnull().sum() > 0:
        df.loc[:, col] = df[col].fillna(df[col].median())
```

```python
for col in categorical_cols:
    if df[col].isnull().sum() > 0:
        df.loc[:, col] = df[col].fillna(df[col].mode()[0])
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, MinMaxScaler
```

```python
# Label Encoding for "Has Car"
label_encoder = LabelEncoder()
df["Has Car"] = label_encoder.fit_transform(df["Has Car"])
df
```

```python
print(df.columns)  # Check the available column names
```

```python
df.columns = df.columns.str.strip()  # Remove extra spaces
```

```python
categorical_cols = ["Education Level", "Job Role", "City"]
existing_cols = [col for col in categorical_cols if col in df.columns]

if existing_cols:  
    df = pd.get_dummies(df, columns=existing_cols, drop_first=True)
else:
    print("One-hot encoding is already done or columns do not exist.")
```

```python
df.head()
```

```python
# Scaling Numerical Features
scaler = MinMaxScaler()
df[["Age", "Salary", "Work Experience (Years)", "Purchase Amount"]] = scaler.fit_transform(
    df[["Age", "Salary", "Work Experience (Years)", "Purchase Amount"]]
)
```

```python
print(df.head())
```

```python
# Convert boolean columns to integer (0/1)
boolean_cols = ["Education Level_High School", "Education Level_Master's", "Education Level_PhD",
                "Job Role_Engineer", "Job Role_Manager", "Job Role_Technician",
                "City_Houston", "City_New York", "City_San Francisco"]

df[boolean_cols] = df[boolean_cols].astype(int)

# Ensure "Has Car" is also properly encoded as integer (if not already)
df["Has Car"] = df["Has Car"].astype(int)

# Display the first few rows of the fully encoded dataset
df.head()
```
